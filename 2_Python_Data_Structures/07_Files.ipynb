# 7.1 Files

## 1. Opening a File

### open()

- returns a "file handle"
- **handle = open("filename", mode)**
- **"filename"** is a string
- **"mode"** is optional, and should be **"r"** if we are planning to read the file, and **"w"** if we are going to write to the file

## 2. What is a Handle?

- **handle** is not an actual file
- (open, read, write, close) -> HANDLE -> mbox.txt

```python
handle = open("mbox.rtf")

print(handle)    # not an actual file
```

## 2. Counting Lines in a File

```python
handle = open("mbox.rtf")
count = 0

for line in handle:
    count = count + 1

print("Line Count:", count)
```

## 3. Reading the Whole File

### .read()

- read the whole file into a **single string**

```python
handle = open("mbox.rtf")

inp = handle.read()
# print(inp)          # prints the entire file

print(inp[:20])
```

## 4. Searching Through a File

```python
handle = open("mbox.rtf")

for line in handle:
    
    if line.startswith("From:"):
        print(line)
```

### Why blank spaces between lines?
- Because each line in the text already has **\n** at the end,  
  and the **print()** statement adds **\n** after each output
  
### To fix this

```python
handle = open("mbox.rtf")

for line in handle:
    line = line.rstrip()    # this line is added
    
    if line.startswith("From:"):
        print(line)
```

### Using "in" to Select Lines

```python
handle = open("mbox.rtf")

for line in handle:
    line = line.rstrip()
    
    if not "@uct.ac.za" in line:
        continue
    
    print(line)
```

2. Write a program that prompts for a file name, then opens that file and reads through the file, looking for lines of the form:

X-DSPAM-Confidence:    0.8475

Count these lines and extract the floating point values from each of the lines and compute the average of those values and produce an output as shown below. Do not use the sum() function or a variable named sum in your solution.
You can download the sample data at http://www.py4e.com/code3/mbox-short.txt when you are testing below enter **mbox.rtf** as the file name.

```python
# Use the file name mbox-short.txt as the file name
fname = input("Enter file name: ")
fh = open(fname)

total = 0
count = 0

for line in fh:
    
    if line.startswith("X-DSPAM-Confidence:") :
        a = line.find("0")
        
        x = line[a:(a + 6)]
        y = float(x)
        
        count = count + 1
        total = total + y
        
        ans = total / count

print("Average spam confidence:", ans)
```
